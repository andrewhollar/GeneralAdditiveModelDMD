```python
#Import Cell
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pygam import LinearGAM, s, f
```

```python
#Load both excel spreadsheets into DataFrames
popplewell = pd.read_excel('data/popplewellData.XLSX', index_col = 0)
otherStudies = pd.read_excel('data/otherStudiesData.XLSX', index_col = 0)
```

```python
#Remove columns with text as data (For Now!)
colWString = ["Author", "Chemistry", "Oligo", "Skip level (reported)", "Exon Malueka Category"]
X = popplewell
for c in colWString:
    X = X.drop(c, 1)
```

```python
#Construct the dataFrames for both X (features), y (target)
y = X['% skip']
X = X.drop("% skip", 1)
```

```python
#Convert X, y to matrices
Xmatrix = X.values
yMatrix = y.values
```

```python
#Create LinearGAM [What does this mean?]

gam = LinearGAM(s(0, n_splines=5) + s(1) + f(2)).fit(Xmatrix,yMatrix)
lam = np.logspace(-3, 5, 5)
lams = [lam] * 3


gam.gridsearch(X, y, lam = lams)
gam.summary()
```

```python
#Plot linearGam [What does this mean?]
for i, term in enumerate(gam.terms):
    if term.isintercept:
        continue

    XX = gam.generate_X_grid(term=i)
    pdep, confi = gam.partial_dependence(term=i, X=XX, width=0.95)

    plt.figure()
    plt.plot(XX[:, term.feature], pdep)
    plt.plot(XX[:, term.feature], confi, c='r', ls='--')
    plt.title(repr(term))
    plt.show()
```
